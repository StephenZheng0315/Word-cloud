```python
import numpy as np
import pandas as pd
from os import path
from PIL import Image
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
import warnings
warnings.filterwarnings("ignore")

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
f = open(u'temp.txt','r').read()
wordcloud = WordCloud(background_color="white",width=1000, height=460, margin=2).generate(f)
# width,height,margin:You can set image properties
# generate : You can do automatic word splitting for all text,
# but it doesn't support Chinese well, please see my next article for Chinese word splitting processing.
# wordcloud = WordCloud(font_path = r'D:\Fonts\simkai.ttf').generate(f)
# You can set the font set with the font_path parameter
# background_color:The parameter is to set the background color, the default color is black
```

```python
plt.figure(figsize=(100, 46), dpi=200)
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```
